```python
import faiss
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('D:/Felicity/4/spotify_data.csv')
print(df.columns)
```

```python
df[['artist_name', 'track_name', 'popularity',
       'year', 'genre', 'danceability', 'energy', 'key', 'loudness', 'mode',
       'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'duration_ms']]
```

```python
random_rows = df.sample(n=10, random_state=1234)
```

```python
features = ['danceability', 'energy', 'key', 'loudness', 'mode', 
            'speechiness', 'acousticness', 'instrumentalness', 
            'liveness', 'valence', 'tempo', 'duration_ms']
```

```python
scaler = StandardScaler()
normal_features = scaler.fit_transform(df[features])
normal32 = normal_features.astype('float32')
index = faiss.IndexFlatL2(normal32.shape[1])
index.add(normal32)
k = 5
xq = scaler.transform(random_rows[features]).astype('float32')
D, I = index.search(xq, k)
```

```python
similar_tracks = df.iloc[I.flatten()]['track_id'].values.reshape(I.shape)
print("Track IDs of similar tracks:\n", similar_tracks)
```
